# Machine Learning (K-Nearest Neighbor)

```python
data = pd.read_csv('ECommerce_Cleansed_KNN.csv', index_col = 'CustomerID')
```

```python
data
```

```python
data.columns[28]
```

```python
# grouping the attributes into the features list
features = list(data.columns[1:128])
```

```python
# use the names of attributes to split them into independent variables X and target variable y
X = data[features]
y = data.iloc[:,0]
```

```python
y
```

```python
#import functions for building the KNN model and evaluation
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.metrics import classification_report,confusion_matrix, recall_score, precision_recall_curve
```

```python
#split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 42)
```

### baseline model

```python
baseline = KNeighborsClassifier()
baseline = baseline.fit(X_train, y_train)
pred_val_test_baseline = baseline.predict(X_test)
#evaluation
print('Confusion Matrix:', '\n',confusion_matrix(y_test,pred_val_test_baseline))
print('Classification Report:', '\n', classification_report(y_test,pred_val_test_baseline))
```

### Grid search CV

```python
# define weight
weights = ['uniform', 'distance']

#define n_neighbours
neighbors = np.arange(5, 65, 5)

#initiate grid
param_grid  = dict(weights=weights, n_neighbors=neighbors)
param_grid
```

```python
knn = KNeighborsClassifier()
grid_search = GridSearchCV(knn, param_grid, cv=10)
grid_search.fit(X_train, y_train)
```

```python
#finding the best parameter
grid_search.best_params_
```

```python
grid_search.best_score_
```

# Evaluation

```python
knn = KNeighborsClassifier(n_neighbors=2, weights='distance')
knn.fit(X_train,y_train)
pred_val_test = knn.predict(X_test)
print('Confusion Matrix:', '\n', confusion_matrix(y_test,pred_val_test))
print('Classication Report', '\n', classification_report(y_test,pred_val_test))
```

### Precision-Recall Curve

```python
test_y_array = np.array(y_test)
```

```python
predict_baseline_proba = baseline.predict_proba(X_test)
predict_proba = knn.predict_proba(X_test)
```

```python
precision1, recall1, threshold1 = precision_recall_curve(test_y_array, predict_baseline_proba[:,1])
precision2, recall2, threshold2 = precision_recall_curve(test_y_array, predict_proba[:,1])
```

```python
plt.plot(precision1,recall1, label='baseline')
plt.plot(precision2,recall2, label='revised model') #cvgrid new model
plt.legend()
```

### ROC / AUC Curve

```python
fpr1, tpr1, threshold3 = metrics.roc_curve(test_y_array, predict_baseline_proba[:,1])
fpr2, tpr2, threshold4 = metrics.roc_curve(test_y_array, predict_proba[:,1])
```

```python
plt.plot(fpr1,tpr1, label='baseline') #baseline
plt.plot(fpr2,tpr2, label='revised model') #cvgrid new model
plt.legend()
```

```python
# auc of the baseline model
metrics.auc(fpr1, tpr1)
```

```python
# auc of the revised model
metrics.auc(fpr2, tpr2)
```
