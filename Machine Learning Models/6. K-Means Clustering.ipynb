# Data Initialization

```python
# Import Libraries

import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.cluster import KMeans

# load dataset

data = pd.read_csv('ECommerce_Cleansed_Clustering.csv', index_col = 'CustomerID')
```

```python
data.head()
```

# Data Preparation

```python
# define X features

features = ['CityTier','NumberOfDeviceRegistered','SatisfactionScore','NumberOfAddress','Complain','CouponUsed',
            'OrderCount','Tenure','HourSpendOnApp','WarehouseToHome','DaySinceLastOrder',
            'OrderAmountHikeFromlastYear','CashbackAmount']
X = data[features]
```

```python
# pre-processing of X

X
```

# K-Means Clustering

```python
# set number of cluster to be 3

kmeansmodel = KMeans(n_clusters=3,n_init=10,max_iter=100)
```

```python
# fit data

kmeansmodel.fit(X)
```

```python
# centroids

centroids = kmeansmodel.cluster_centers_
centroids
```

```python
# number of interation

interation = kmeansmodel.n_iter_
interation
```

```python
# sum of squared distances of samples to their closest cluster center

SSD = kmeansmodel.inertia_
SSD
```

```python
# predict cluster number

cluster = kmeansmodel.predict(X)
cluster
```

```python
X['Cluster'] = cluster
```

```python
X.groupby(['Cluster']).mean()
```

```python
X_new = X[['Cluster','NumberOfAddress','CouponUsed','OrderCount','DaySinceLastOrder']]
X_new.groupby(['Cluster']).mean()
```
